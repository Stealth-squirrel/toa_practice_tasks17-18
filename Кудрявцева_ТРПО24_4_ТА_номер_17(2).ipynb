```python
import itertools
```

```python
def opt_order(r):
    best_s = -1
    best_orders = []
    for perm in itertools.permutations(r):
        s = sum(100 * x ** (t+1) for t, x in enumerate(perm))
        if s > best_s:
            best_s = s
            best_orders = [perm]
        elif s == best_s:
            best_orders.append(perm)
    return best_orders, best_s
```

```python
r = [3/4, 1/2, 1/100]   # коэффициенты снижения
best_orders, best_rev = opt_order(r)
print("Оптимальный порядок:", best_orders)
print("Максимальная выручка:", best_rev)
```
