```python
def subsequence(s_big, s_small):
    i = 0  # индекс элемента s_big
    j = 0  # индекс элемента s_small
    n, m = len(s_big), len(s_small)

    while i < n and j < m:
        if s_big[i] == s_small[j]:
            j += 1
        i += 1

    return j == m
```

```python
# пример использования
s_big = ["купить Amazon", "купить Yahoo", "купить eBay", "купить Yahoo", "купить Yahoo", "купить Oracle"]
s_small = ["купить Yahoo", "купить eBay", "купить Yahoo", "купить Oracle"]
print(subsequence(s_big, s_small))

s_small2 = ["купить Yahoo", "купить Oracle", "купить eBay", "купить Yahoo"]
print(subsequence(s_big, s_small2))
```
